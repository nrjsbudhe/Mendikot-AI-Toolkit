```python
import matplotlib.pyplot as plt

# Sample data
data_1 = [95, 100, 105, 110, 115]  # Values for variable 1
data_2 = [190, 195, 200, 205, 210]  # Values for variable 2

# Calculate bin range
bin_min = min(min(data_1), min(data_2))
bin_max = max(max(data_1), max(data_2))
bin_width = 10  # Adjust as needed

# Define bins
bins_1 = range(bin_min, bin_max + bin_width, bin_width)
bins_2 = [bin + bin_width / 2 for bin in bins_1]

# Plot histogram
plt.hist(data_1, bins=bins_1, alpha=0.5, label='Variable 1', edgecolor='black')
plt.hist(data_2, bins=bins_2, alpha=0.5, label='Variable 2', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Comparison Histogram')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the first set of bars
categories = ['Category 1', 'Category 2', 'Category 3']
values1 = [50, 70, 90]  # Example values for the first set

# Data for the second set of bars
values2 = [60, 80, 85]  # Example values for the second set

# Define the width of the bars
bar_width = 0.35

# Define the x locations for the groups
x = range(len(categories))

# Plotting the bars for each set of data
plt.bar(x, values1, width=bar_width, label='Set 1')
plt.bar([i + bar_width for i in x], values2, width=bar_width, label='Set 2')

# Adding labels, title, and legend
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Comparative Bar Graph')
plt.xticks([i + bar_width / 2 for i in x], categories)
plt.legend()

# Show the plot
plt.show()
```
